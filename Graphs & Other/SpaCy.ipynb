```python
from transformers import pipeline

# Carica il modello preaddestrato per l'analisi del sentiment in italiano
sentiment_analysis = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Frase di esempio in italiano
example_sentence_italian = "Io odio e disprezzo spaCy!"

# Esegui l'analisi del sentiment
result_italian = sentiment_analysis(example_sentence_italian)

# Stampa il risultato
print(f"Sentimento della frase in italiano: {result_italian[0]['label']} con una confidenza del {result_italian[0]['score']}")
```

```python
def sentiment_analysis(text):
    classifier = pipeline('sentiment-analysis', model="nlptown/bert-base-multilingual-uncased-sentiment")
    result = classifier(text)[0]

    sentiment = result['label']
    score = result['score']

    return sentiment, score

if __name__ == "__main__":
    frase = input("Inserisci una frase per l'analisi del sentiment: ")
    risultato, punteggio = sentiment_analysis(frase)
    print(f"Sentiment: {risultato}")
    print(f"Punteggio: {punteggio}")
```

```python
import spacy
from textblob import TextBlob
```

```python
def sentiment_analysis(text):
    nlp = spacy.load("it_core_news_sm")
    doc = nlp(text)

    # Estrai il testo dalla doc e crea un oggetto TextBlob per l'analisi del sentiment
    text_blob = TextBlob(doc.text)

    # La polarità è un valore compreso tra -1 e 1, dove -1 indica un sentimento negativo, 0 neutro e 1 positivo.
    polarity = text_blob.sentiment.polarity

    if polarity > 0:
        sentiment = "Positivo"
    elif polarity < 0:
        sentiment = "Negativo"
    else:
        sentiment = "Neutro"

    return sentiment, polarity

if __name__ == "__main__":
    frase = input("Inserisci una frase per l'analisi del sentiment: ")
    risultato, polarita = sentiment_analysis(frase)
    print(f"Sentiment: {risultato}")
    print(f"Polarità: {polarita}")
```
