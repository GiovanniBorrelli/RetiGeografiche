```
import pandas as pd
import matplotlib.pyplot as plt

# Leggi i dati dal file Excel
df = pd.read_excel('input.xlsx')

# Grafico delle emozioni
plt.figure(figsize=(8, 5))
df['Emozione'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuzione delle emozioni')
plt.xlabel('Emozione')
plt.ylabel('Conteggio')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Grafico dei sentimenti
plt.figure(figsize=(8, 5))
df['Sentimento'].value_counts().plot(kind='bar', color='salmon')
plt.title('Distribuzione dei sentimenti')
plt.xlabel('Sentimento')
plt.ylabel('Conteggio')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Converti i valori della colonna "Frasi" in stringhe
df['Frasi'] = df['Frasi'].astype(str)

# Concatena tutte le frasi in una singola stringa
text = ' '.join(df['Frasi'])

# Crea il word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Visualizza il word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Converti i valori della colonna "Frasi" in stringhe
df['Frasi'] = df['Frasi'].astype(str)

# Concatena tutte le frasi in una singola stringa
text = ' '.join(df['Frasi'])

# Crea il word cloud con uno stile pi bello
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis',
                      max_words=100, contour_color='steelblue', contour_width=1).generate(text)

# Visualizza il word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
